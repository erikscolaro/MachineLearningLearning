```python
import numpy as np
import matplotlib 
import matplotlib.pyplot as plot

filename = "./iris.csv"
```

```python
dataset = []
labelset = []
labels = {
	"Iris-setosa":0,
	"Iris-versicolor":1,
	"Iris-virginica":2
}

with open(filename, "r") as file:
	for line in file:
		fields = line.split(",")
		dataset.append(fields[:4])
		labelset.append(labels[fields[4].strip()])
		
dataset = np.array(dataset, dtype=np.float64).reshape((150,4))
labelset = np.array(labelset, dtype=np.int8).reshape((150, 1))
```

```python
plotdim = [1,4]
plot.figure(figsize=(8,2))
s0 = dataset[labelset.flatten() == 0]
s1 = dataset[labelset.flatten() == 1]
s2 = dataset[labelset.flatten() == 2]

for i in range(4):
    plot.subplot(*plotdim, i + 1)
    plot.hist(s0[:, i], bins= 15, alpha = 0.5, label = "Setosa", density=True)
    plot.hist(s1[:, i], bins= 15, alpha = 0.5, label = "Versicolor", density=True)
    plot.hist(s2[:, i], bins= 15, alpha = 0.5, label = "Virginica", density=True)
    
plot.show()
```

```python
plotdim = [4,4]
plot.figure(figsize=(24,16))
s0 = dataset[labelset.flatten() == 0]
s1 = dataset[labelset.flatten() == 1]
s2 = dataset[labelset.flatten() == 2]

for i in range(4):
    for j in range(4):
        plot.subplot(*plotdim, j*4+i+1)
        if (i==j):
            plot.hist(s0[:, i], bins= 15, alpha = 0.5, label = "Setosa", density=True)
            plot.hist(s1[:, i], bins= 15, alpha = 0.5, label = "Versicolor", density=True)
            plot.hist(s2[:, i], bins= 15, alpha = 0.5, label = "Virginica", density=True)
            plot.legend()
        else:
            plot.scatter(s0[:, i], s0[:, j], alpha=0.5, label="Setosa", marker='o')
            plot.scatter(s1[:, i], s1[:, j], alpha=0.5, label="Versicolor", marker='^')
            plot.scatter(s2[:, i], s2[:, j], alpha=0.5, label="Virginica", marker='s')
            plot.legend()
```

```python
mean = dataset.mean(1, keepdims=True)
dataset_centrated = dataset-mean

covariance_matrix = dataset_centrated@dataset_centrated.T / float(dataset.shape[1])

var = dataset.var(1)
std = dataset.std(1)

print(covariance_matrix.shape)
print(var.shape)
print(std.shape)
```
